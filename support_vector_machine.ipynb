```python
# import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.svm import SVC
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("data/data.csv")
```

```python
#load the dataset information
df.info()
```

```python
#load the dataset first five row head
df.head()
```

```python
# drop irrelevant columns
df.drop(columns=["id", "Unnamed: 32"], inplace=True)
df.head()
```

```python
#Encode target to classification binary
df['diagnosis'] = df['diagnosis'].map({'M': 1, 'B': 0})
df.head()
```

```python
df.info()
```

```python
#load features and target 
X = df.drop('diagnosis', axis=1)
y = df['diagnosis']
```

```python
#scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
#dimensional reduction plotty
pca = PCA(n_components=2)
x_pca =pca.fit_transform(X_scaled)
```

```python
#Split the data.
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2,random_state=42)
```

```python
#train SVM model

svm_model = SVC(kernel='rbf', C=1.0, gamma='scale')
svm_model.fit(X_train, y_train)
```

```python
#predict and evaluate
y_pred = svm_model.predict(X_test)
print(classification_report(y_test, y_pred))
```

```python
#visualization
#create a 2D data using the PCA
X_test_pca = pca.transform(X_test)
```

```python
#Create DataFrame for the ploting 

plot_df = pd.DataFrame(X_test_pca, columns=['PCA1', 'PCA2'])
plot_df['Prediction'] = y_pred
plot_df['Actual'] = y_test.values
```

```python
#scatter plot visualization
plt.figure(figsize=(8, 6))
sns.scatterplot(data=plot_df, x='PCA1', y='PCA2', hue='Prediction', style='Actual', palette='coolwarm', s=100)
plt.title("SVM Classification Report")
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.legend(title='Prediction / Actual')
plt.grid(True)
plt.tight_layout()
```
